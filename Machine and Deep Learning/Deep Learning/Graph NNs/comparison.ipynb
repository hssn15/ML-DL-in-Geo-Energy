```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import matplotlib
matplotlib.style.use('ggplot')
```

```python
r2_p_gat = np.genfromtxt(r"C:\Users\YUSIFOH\NNs\GNN\comparison data\r2_p_gat.mat")
r2_s_gat = np.genfromtxt(r"C:\Users\YUSIFOH\NNs\GNN\comparison data\r2_s_gat.mat")

r2_p_gconv = np.genfromtxt(r"C:\Users\YUSIFOH\NNs\GNN\comparison data\r2_p_gconv.mat")
r2_s_gconv = np.genfromtxt(r"C:\Users\YUSIFOH\NNs\GNN\comparison data\r2_s_gconv.mat")

r2_p_gcn = np.genfromtxt(r"C:\Users\YUSIFOH\NNs\GNN\comparison data\r2_p_gcn.mat")
r2_s_gcn = np.genfromtxt(r"C:\Users\YUSIFOH\NNs\GNN\comparison data\r2_s_gcn.mat")
```

```python
plt.figure(figsize=(6, 6), dpi = 100)
plt.fill_between(np.log(np.arange(len(r2_s_gconv))), r2_s_gconv, color="g",   alpha=1, label='GraphConv')
plt.fill_between(np.log(np.arange(len(r2_s_gat))),   r2_s_gat, color="r",     alpha=1, label='GAT')
plt.fill_between(np.log(np.arange(len(r2_s_gcn))),   r2_s_gcn, color="b",     alpha=1, label='GCNconv')
plt.grid()
plt.legend()
plt.title("R2 Score Area Plot | Saturation")
plt.savefig(r"C:\Users\YUSIFOH\NNs\GNN\DATA DRIVEN\R2 Score Area Plot Saturation.png")
plt.show()
```
